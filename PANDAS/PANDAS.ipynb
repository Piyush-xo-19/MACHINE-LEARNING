Series

ACESSING AND SLICING 

```python
#print(df.columns)
#print(df.property_name) # can be used to get values of a particular column 
#print(df["property_name"])# another method to call the value of a column 
print(df[["property_name","price"]])# method to call multiple colums at a time 
```

```python
"""
acessing rows :-
  so to acess rows we have 2 methods one is .loc and send one is .iloc loc stands for label based acessing function and 
  .iloc is index label based function 
"""
print(df.loc[0:10,["property_name","price"]]) 
'''
 so 1 here 0:10 is an operation used to call the rows and  list of [column name ] is passed in this  
 in .loc we have to pass the name of the colulmns if labels are used in in rows then we have to also pass the name
 of rows and apply operation on them '''
 # exmaple :-
animals=pd.DataFrame({"tiger":[30,40,50],"cows":[70,80,99],"elephant":[85,45,65]},index=[2017,2018,2019])
print(animals)
# so now i want  to get the the 2017 entry 
print(animals.loc[2017:2018,"tiger":"elephant"]) # so in 1 part we have done slicing in rows and in 2 part we have done slicing in column 
```

```python
#using iloc method in this method instead of adding str we  uses the index of both rows and columns 

#print(df.head())
print(df.iloc[0:10:2,1:5])

'''so  you can see this in the .iloc method we have used the index of the data which we want in the 1 part we have done slicing in  row
and in second part we have done slicing in colummns'''

print(df.iloc[0:10,[0,10,5]])# here we want to acess these 3 colums 
```
