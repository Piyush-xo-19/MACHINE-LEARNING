** BUILDING LINEAR REGRESSION **

```python
import numpy as np

class LinearRegression:
    def __init__(self, learning_rate=0.001, no_of_iterations=1000):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.no_of_iterations):
            y_pred = np.dot(X, self.weights) + self.bias

            # Gradient calculation (
            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)

            # Updating parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 6, 9, 12, 15])  # Example: y = 3x

# ---------------------------
# Train Model
# ---------------------------
model = LinearRegression(learning_rate=0.01, no_of_iterations=1000)
model.fit(X, y)
predictions = model.predict(X)

# ---------------------------
# Plotting
# ---------------------------
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predictions, color='red', label='Regression Line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)
plt.show()
```

ASSIGNING THE COLUMNS TO X AND Y 

```python
X=df.iloc[:,0:1]  #this can be done in the same way by X=df["name of the columns"]
Y=df.iloc[:,-1]   # y=df["package"]
```

DISTRIBUTING DATASETS IN TEST TRAIN 

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state=42)
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
print(X_test)

```

```python
predictions=lr.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
MAE=mean_absolute_error(Y_test,predictions)
MSE=mean_squared_error(Y_test,predictions)
r =r2_score(Y_test,predictions)
print("MEAN ABSOLOUTE ERROR IS ",MAE)
print("Mean squared error is ",MSE)
print("R^2 ERROR = ",r)
```
